```python
import numpy as np
import pandas as pd

df = pd.read_csv(r'C:\NYU\Data Science in Quantitative Finance\Short Term Project\data\SPX_OHLC.csv')
df.head()
```

```python
df['Close'].size
n = df['Close'].size    # set n = # of daily observations
```

```python
# Calculate natural log of Opening prices for trade date t & t-1:
df['O_t'] = np.log(df['Open'])
df['O_t-1'] = np.log(df['Open'].shift(1))

# Calculate log difference of Opening prices from t to t-1: 
df['o_t'] = df['O_t'] - df['O_t-1'] 
df.head()
```

```python
# Calculate natural log of Closing prices for trade date t & t-1:
df['C_t'] = np.log(df['Close'])
df['C_t-1'] = np.log(df['Close'].shift(1))

# Calculate log difference of Closing prices from t to t-1: 
df['c_t'] = df['C_t'] - df['C_t-1'] 
df.head()
```

```python
# Calcualte average log difference in daily Opening prices
o_bar = (1/n) * np.sum(df['o_t'])

# Calcualte average log difference in daily Closing prices
c_bar = (1/n) * np.sum(df['c_t'])

o_bar, c_bar
```

```python
# Calculate daily variance of Opening & Closing log prices from their sample averages
df['o_dv'] = np.square(df['o_t'] - o_bar)
df['c_dv'] = np.square(df['c_t'] - c_bar)

df.head()
```

```python
V_o = (1 / (n - 1)) * np.sum(df['o_dv'])
V_c = (1 / (n - 1)) * np.sum(df['c_dv'])

V_o, V_c
```

```python
# Calculate natural log of High & Low prices for trade date t:
df['H_t'] = np.log(df['High'])
df['L_t'] = np.log(df['Low'])

df.head()
```

```python
V_rs = (1 / n) * np.sum(((df['H_t'] - df['O_t'])*(df['H_t'] - df['C_t'])) + ((df['L_t'] - df['O_t'])*(df['L_t'] - df['C_t'])))
V_rs
```

```python
k = 0.34 / (1.34 + ((n + 1) / (n - 1)))
            
V_real = np.sqrt((V_o + (k*V_c) + ((1 - k)*V_rs)))
V_real
```

```python
V_est = np.sqrt(252) * V_real
V_est
```

```python
df_VIX = pd.read_csv(r'C:\NYU\Data Science in Quantitative Finance\Short Term Project\data\VIX.csv')
df_VIX.head()
```

```python
df_VRP = pd.concat([df_VIX['Date'], df_VIX['Close'] - (V_est * 100)], axis=1, keys=['Date', 'VRP'])
df_VRP = df_VRP.set_index(['Date'])

df_VRP.head()
```

```python
df_VRP.to_csv(r'C:\NYU\Data Science in Quantitative Finance\Short Term Project\data\VRP.csv')
```
