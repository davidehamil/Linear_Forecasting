```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv(r'E:\NYU MathFin\courses\data science in quant finance\short project\data\SIM.csv',parse_dates=[0],usecols=[0])
```

```python
years = np.arange(1990,2018)
btw_dates = []

for year in years:
    may_date = df[df['Date']>= datetime.datetime(year,5,1)].iloc[1,0]
    oct_date = df[df['Date']>= datetime.datetime(year,10,1)].iloc[14,0]
    btw_date = df[(df['Date']>= may_date) & (df['Date']<= oct_date)].iloc[:,0].tolist()
    btw_dates = btw_dates + btw_date
    
btw_dates
```

```python
df['SIM'] = 0

for i in range(len(df)):
    fut_dates = df['Date'].iloc[i+1:i+130].tolist()
    itsct = list(set(fut_dates).intersection(set(btw_dates)))
    df['SIM'].iloc[i] = (len(itsct))/130.
```

```python
df = df[(df['Date'] >= datetime.datetime(1990,1,1)) & (df['Date'] <= datetime.datetime(2017,9,29))]
df.set_index(['Date'])
df.to_csv('SIM.csv')
```
